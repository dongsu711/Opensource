```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv('CAvideos.csv')
df1.head()
```

```python
df1.index = pd.to_datetime(df1.index)
```

```python
fig = plt.figure(figsize=(20, 12))
ax = fig.add_subplot(1,1,1)

ax.plot(df1['likes'], label='df1', color='b')
ax.plot(df1['dislikes'].iloc[1689:], label='df1', color='r')

ax.set_title('Youtbe likes & dislikes', fontsize=20)
ax.set_ylabel('likes&dislikes', fontsize=15)

ax.legend(fontsize=12, loc='best')
plt.show()
```

```python
X = df1["likes"]
y = df1["dislikes"]
plt.plot(X,y,'o')
plt.show()
line_fitter = LinearRegression()
line_fitter.fit(X.values.reshape(-1,1),y)
line_fitter.predict([[1000]])
```

```python
line_fitter.coef_
```

```python
line_fitter.intercept_
```

```python
plt.plot(X,y,'o')
plt.plot(X,line_fitter.predict(X.values.reshape(-1,1)))
plt.show()
```
